#### __Basic Example__ 
-----
* The code below graphs the function f(x) = x^2
  * __plt.grid__ puts a grid behind our graph. I usually set the grid color to '#CCCCCC', the linewidth to 0.5, and the linestyle to '--'. This is not related to __np.linspace__!
  * __np.linspace(-3, 3, 20)__ divides the x-axis of the graph into 20 spaces starting from -3 to 3. The values of f(x) = x^2 will be plotted at these intervals on the array. Your graph may look as smooth as PS3 Lara Croft or as jank as PS1 Lara Croft depending on how many spaces you dived x-axis into.
  * We've also printed the values of our x-axis generated by __np.linspace__ as an array. Note than any functions graphed must have the same shape as this array, or else your graph won't print.
  * I usually set the __color__ of the graph to match the colored box it I'm putting it under in my notes.

```python
x = np.linspace(-3, 3, 20)
y = x**2
```

```python
x
```

```python
plt.grid(color = '#CCCCCC', linewidth = 0.5, linestyle = '--')
plt.plot(x, y, color = '#f1c232', linestyle = '-')
```

#### __Basic Matplotlib Commands__ 
-----
* Using the same __x = np.linspace(-3, 3, 20)__ as in the yellow box, we again graphed the function f(x) = x^2

* The code below takes the function f(x) = x^2 and graphs it two ways"

  * x a as a function of y (light blue dashed)

__===========================================================================__
#### __Introduction to the Object-Oriented Method__ 
-----
* Below, we again graph the function f(x) = x^2. This time we're create a figure in which we we can find the graph.
  * This time, we included no line style and only the marker '*'

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.8])
plt.grid(color = '#CCCCCC', linewidth = 0.5, linestyle = '--')
axes.plot(x, y, color = '#674ea7', marker = '*', linestyle = 'none')
```

------
* Nowe we're going to plot two graphs of the function f(x) = x^2 on the same canvas. 
  * __Larger Graph__ will be y as a function of x
  * __Smaller Graph__ will be x as a function of y
* The size of of the overall figure will be 10 x 8
* You'll see that the graphs overlap which make it hard to read. We'll fix this later.
* We also removed the grid because two differently sized grids imposed on each other would get hella confusing.

```python
fig = plt.figure(figsize = (10, 8))

axe1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe1.plot(x, y, color = '#674ea7', marker = 'o', lw = 2)
axe1.set_title('Larger Graph: y as a function of x')

axe2 = fig.add_axes([0.3, 0.1, 0.2, 0.3])
axe2.plot(y, x, color = '#b4a7d6', marker = '>', lw = 2)
axe2.set_title('Smaller Graph: x as a function of y')
```

------
* Since the last figure was too big and wieldy, we removed the __figsize__ to set it back to the smaller default
* We adjusted the axes numbers on the smaller graph to fix that overlap.

```python
fig = plt.figure()

axe1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe1.plot(x, y, color = '#674ea7', marker = 'o')
axe1.set_title('Larger Graph')

axe2 = fig.add_axes([0.4, 0.4, 0.25, 0.25])
axe2.plot(y, x, color = '#b4a7d6', marker = '>')
axe2.set_title('Smaller Graph')
```

#### __subplots()__ 
-----
* The __plt.subplots()__ object willwill ac as a more automatic axis maanager
* See an example of a basic use case below for the function f(x) = x^2:

```python
# Use similar to plt.figure() except use tuple unpacking to graph fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, color = '#ff0000')
plt.grid(color = '#CCCCCC', linewidth = 0.5, linestyle = '--')
axes.set_xlabel = ('x-label')
axes.set_ylabel = ('y-label')
axes.set_title('f(x) = x^2')
```

-----
* Then you can specify the number of rows and columns when creating the __subplots()__ object:

```python
# Empy canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (12, 5))
```

```python
# Axes is an array of axes to plot on
axes
```

-----
* We can iterate though this array:

```python
for ax in axes:
    ax.plot(x, y, '#ff0000')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    ax.set_title('Repeating Title')

# Display the figure object
fig
```

#### __Titles, Axis Labels, and Legends__ 
-----
* Now that we ahve covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how to add the following to a figure:
  * titles
  * axis labels
  * legends

-----
* __Figure Titles__
<br>
  A title can be added to each axis instance in a figure. To set the title, use the __set_title___ method in the axes instance:

```python
ax.set_title('Insert Title Here')
```

-----
* __Axis Labels__
<br>
  Similarly, with the methods __set_xlabel__ and __set_ylabel__, we can set the labels of the x and y axes.

```python
ax.set_xlabel('x')
ax.set_ylabel('y')
```

-----
* __Legends__
  * You can see the __label = 'label text'__ keyword argument when plots or other objects are added to the figure, and then using the __legend__ method without arguments to add the legend to the figure.
  * If you scroll down, you can see how changing the paramater after __loc__ can change the position of the legend. The default is always 0 or 'best.'
    * 0 = 'best'
    * 1 = 'upper right'
    * 2 = 'upper left'
    * 3 = 'lower left'
    * 4 = 'lower right'
    * 5 = 'right'
    * 6 = 'center left
    * 7 = 'center right'
    * 8 = 'lower center'
    * 9 = 'upper center'
    * 10 = 'center'

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
plt.grid(color = '#CCCCCC', linewidth = 0.5, linestyle = '--')
ax.plot(x, x**2, color = '#f1c232', marker = '*', linestyle = 'none', label = 'y = x^2')
ax.plot(x, x**3, color = '#b4a7d6', marker = 'none', linestyle = '--', label = 'y = x^3')
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
plt.grid(color = '#CCCCCC', linewidth = 0.5, linestyle = '--')
ax.plot(x, x**2, color = '#f1c232', marker = '*', linestyle = 'none', label = 'y = x^2')
ax.plot(x, x**3, color = '#b4a7d6', marker = 'none', linestyle = '--', label = 'y = x^3')
ax.legend(loc = 9)
```
